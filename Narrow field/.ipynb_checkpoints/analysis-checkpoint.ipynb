```python
from snAPI.Main import *
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
class HH_T3:
    def __init__(self,sn,f,filename):
        self.f=f
        self.Tsync=1e9/f #Time unit is ns
        sn.getDevice()
        sn.getFileDevice(filename)
        sn.initDevice(MeasMode.T3)
        sn.raw.measure(120)
        self.data = sn.raw.getData()
        self.t_resol=sn.deviceConfig['Resolution']/1000
        self.dTimes=sn.raw.dTime_T3(self.data)*self.t_resol
        self.channels=sn.raw.channel(self.data)
        self.nsync=sn.raw.nSync_T3(self.data)
        self.step=np.where(self.channels==64, 1, 0)
        self.cumulative=np.cumsum(self.step)
        self.absTimes=(self.nsync+self.cumulative*1024)*self.Tsync+self.dTimes
        self.df=pd.DataFrame(np.array([self.dTimes,self.absTimes,self.channels,self.nsync]).T,columns=['dTimes','absTimes','Ch','nSync'])
        sn.closeDevice()
    def OnChannel(self,ch):
        return self.df[self.df['Ch']==ch]
        
    
```

```python
sn=snAPI()
```
